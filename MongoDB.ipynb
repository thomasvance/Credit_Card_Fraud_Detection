```python
import pymongo
import pandas as pd 
from pymongo import MongoClient
```

```python
# Read the CSV file
data_path = 'Merged_card_user.csv'  
df = pd.read_csv(data_path, low_memory=False)

df
```

```python
# Select only the relevant columns
filtered_df = df[["Person", "Yearly Income - Person", "Total Debt"]]
print(filtered_df)
```

```python
print(df.dtypes)
```

```python

# Correct data types
df["Current Age"] = df["Current Age"].astype(int)
df["Retirement Age"] = df["Retirement Age"].astype(int)
df["Birth Year"] = df["Birth Year"].astype(int)
df["Birth Month"] = df["Birth Month"].astype(int)
df["Latitude"] = df["Latitude"].astype(float)
df["Longitude"] = df["Longitude"].astype(float)
df["Per Capita Income - Zipcode"] = df["Per Capita Income - Zipcode"].str.replace("[$,]", "", regex=True).astype(float)
df["Yearly Income - Person"] = df["Yearly Income - Person"].str.replace("[$,]", "", regex=True).astype(float)
df["Total Debt"] = df["Total Debt"].str.replace("[$,]", "", regex=True).astype(float)
df["Num Credit Cards"] = df["Num Credit Cards"].astype(int)
df["CARD INDEX"] = df["CARD INDEX"].astype(int)
df["Credit Limit"] = df["Credit Limit"].str.replace("[$,]", "", regex=True).astype(float)
```

```python
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
df['Apartment'].fillna("N/A", inplace=True)
```

```python
print(df.isnull().sum())
```

```python
print(df.columns)
```

```python
#fix any extra unseen spacing
df.rename(columns=lambda x: x.strip(), inplace=True)
```

```python
#Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/") 

#Create DataBase and  Collection:
db = client["Card_fraud_db"]  
collection = db["card_users"]  

#Convert DataFrame to dictionary format for MongoDB
data_dict = df.to_dict("records")

# Insert data into MongoDB collection
collection.insert_many(data_dict)

print("data successfully inserted")
```

```python
# Verify that retrieve data from Mongo Db
retrieved_data = collection.find().limit(10)
retrieved_df = pd.DataFrame(list(retrieved_data))
retrieved_df.head()
```
