```python

# Function to generate random time in 24-hour format
def generate_random_time():
    hour = random.randint(6, 20) if random.random() < 0.8 else random.randint(20, 23)  # 80% 6 AM - 8 PM, 20% 8 PM - 6 AM
    minute = random.randint(0, 59)
    second = random.randint(0, 59)
    return f"{hour:02d}:{minute:02d}:{second:02d}"

# Function to generate transaction amount
def generate_amount():
    if random.random() < 0.05:  # 5% chance of being between $100 and $500
        return round(random.uniform(100, 500), 2)
    else:  # Normal transaction under $100
        return round(random.uniform(1, 99), 2)
```
