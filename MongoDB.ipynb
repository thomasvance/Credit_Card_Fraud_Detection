```python
import pymongo
import pandas as pd 
```

```python
# Read the CSV file
data_path = 'Merged_card_user.csv'  
df = pd.read_csv(data_path, low_memory=False)

#Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/") 

#Create DataBase and  Collection:
db = client["Card_fraud_db"]  
collection = db["card_users"]  

#Convert DataFrame to dictionary format for MongoDB
data_dict = df.to_dict("records")

# Insert data into MongoDB collection
collection.insert_many(data_dict)

print("data successfully inserted")
```

```python
# Verify that retrieve data from Mongo Db
retrieved_data = collection.find().limit(10)
retrieved_df = pd.DataFrame(list(retrieved_data))
retrieved_df.head()
```
